```python
# Define the size of each segment
segment_size = 1000

# Calculate the total number of segments
num_segments = len(df) // segment_size

# Create an empty DataFrame to store label distribution for each segment
label_distribution = pd.DataFrame()

# Calculate the label proportion for each segment
for i in range(num_segments):
    segment = df['Label'][i * segment_size: (i + 1) * segment_size]
    segment_counts = segment.value_counts(normalize=True)  # Calculate proportion
    segment_counts = segment_counts.to_frame().T  # Transpose for concatenation
    label_distribution = pd.concat([label_distribution, segment_counts], ignore_index=True)

# Handle remaining data by adding an extra segment if necessary
if len(df) % segment_size != 0:
    segment = df['Label'][num_segments * segment_size:]
    segment_counts = segment.value_counts(normalize=True)
    segment_counts = segment_counts.to_frame().T  # Transpose for concatenation
    label_distribution = pd.concat([label_distribution, segment_counts], ignore_index=True)

# Fill NaN values with 0 to indicate that the label is absent in that segment
label_distribution = label_distribution.fillna(0)

# Plot stacked bar chart
plt.figure(figsize=(12, 6))
bottom = None  # Initial value for bottom parameter in stacked bar
for label in label_distribution.columns:
    plt.bar(label_distribution.index, label_distribution[label], label=f'Label {label}', bottom=bottom, alpha=0.7)
    bottom = (bottom + label_distribution[label]) if bottom is not None else label_distribution[label]

plt.xlabel('Segment Index')
plt.ylabel('Proportion')
plt.title('Stacked Label Distribution by Segment in the Dataset')
plt.legend(title='Labels')
plt.show()
```

```python
df1_analyze = df.drop(columns=['Destination IP', 'Source IP', 'Flow ID', 'Fwd Header Length.1', 'Timestamp'])

# Drop any rows with nulls
print(f"Dropping {df1_analyze.isnull().any(axis=1).sum()} rows with null value...")
df1_analyze = df1_analyze.drop(index=df1_analyze[df1_analyze.isnull().any(axis=1)].index)

# Drop any rows with infinity
print(f"Dropping {df1_analyze.isnull().any(axis=1).sum()} rows with infinity value...")
df1_analyze = df1_analyze.drop(index=df1_analyze[df1_analyze.isin([np.inf, -np.inf]).any(axis=1)].index)

# Dropping columns with only 1 value
print(f"Dropping {len(df_one_values)} columns with only 1 value...")
df1_analyze = df1_analyze.drop(columns=df_one_values)

# Change flag columns to boolean type
to_bool = []
for x in df1_analyze.columns:
    if "Flag" in x:
        to_bool.append(x)
print(f"Changing {len(to_bool)} columns to boolean...")
df1_analyze[to_bool] = df1_analyze[to_bool].astype('bool')

# Encode Protocol become categorical
print("Changing Protocol value to categorical with one hot encoding")
df1_analyze = pd.get_dummies(df1_analyze, columns=['Protocol'])


# Changing Label to Benign = 0 and DDoS = 1
df1_analyze['Label'] = df1_analyze['Label'].apply(lambda x: 1 if x == "DDoS" else 0)
df1_analyze['Label'] = df1_analyze['Label'].astype('bool')
```

```python
iterations = corr.columns
variance = df1_analyze.select_dtypes('number').var()
mask = np.triu(corr)
columns_to_drop = set()
print("High Linear Correlations\n")
for i in range(len(iterations)):
    for j in range(i):
        if abs(corr.iloc[i,j]) >= 0.8:
            print(f"{iterations[i]} v {iterations[j]}: {corr.iloc[i,j]}\n")
            if variance[iterations[i]] > variance[iterations[j]]:
                columns_to_drop.add(iterations[j])
            else:
                columns_to_drop.add(iterations[i])
        else:
            mask[i,j] = 1

print(f"{len(columns_to_drop)} Columns to drop:\n")
print(columns_to_drop)            
```

                break
        if graph_counter >= nGraphShown:
                break

```python
import pandas as pd
import numpy as np
import numpy as np
import warnings
np.warnings = warnings
from sklearn.preprocessing import StandardScaler, PowerTransformer, LabelEncoder
from sklearn.model_selection import train_test_split, KFold
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
import joblib
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

class NetworkFeatureProcessor:
    """Class for processing network traffic features"""
    
    def __init__(self):
        self.label_encoder = LabelEncoder()
        self.scaler = StandardScaler()
        self.power_transformer = PowerTransformer()
    
    def process_features(self, df):
        """Process and transform network traffic features"""
        df = df.copy()
        
        # Encode Protocol
        df['Protocol'] = self.label_encoder.fit_transform(df['Protocol'])
        
        # Create feature groups
        port_features = ['Source Port', 'Destination Port']
        packet_length_features = ['Fwd Packet Length Max', 'Fwd Packet Length Min', 
                                'Bwd Packet Length Min', 'Min Packet Length', 'Max Packet Length']
        timing_features = ['Bwd IAT Total', 'Bwd IAT Mean', 'Fwd Packets/s']
        flag_features = ['Fwd PSH Flags', 'SYN Flag Count', 'PSH Flag Count', 
                        'ACK Flag Count', 'URG Flag Count']
        other_features = ['Down/Up Ratio', 'Init_Win_bytes_backward', 
                         'min_seg_size_forward', 'Protocol']
        
        # Calculate additional features
        df['Length_Range'] = df['Max Packet Length'] - df['Min Packet Length']
        df['Flag_Sum'] = df[flag_features].sum(axis=1)
        
        # Combine all features
        feature_columns = (port_features + packet_length_features + timing_features + 
                         flag_features + other_features + 
                         [ 'Length_Range', 'Flag_Sum'])
        
        X = df[feature_columns].copy()
        y = (df['Label'] == 'DDoS').astype(int)  # Assuming 'DDoS' is the attack label
        
        return X, y
    
    def scale_features(self, X):
        """Scale and transform features"""
        # Apply log transformation to highly skewed features
        skewed_features = ['Bwd IAT Total', 'Bwd IAT Mean', 'Fwd Packets/s']
        for feature in skewed_features:
            if feature in X.columns:
                X[feature] = np.log1p(X[feature])
        
        # Standard scaling
        X_scaled = self.scaler.fit_transform(X)
        
        # Power transformation
        X_transformed = self.power_transformer.fit_transform(X_scaled)
        
        return X_transformed

class DDoSDetector:
    def __init__(self):
        self.scaler = StandardScaler()
        self.autoencoder = None
        self.ocsvm = None
        self.iforest = None
        self.threshold = None
        self.metrics = {}
        self.feature_importance = {}
        self.cross_val_results = {}
        self.predictions = {}
        
    def build_autoencoder(self, input_dim):
        """Build autoencoder with architecture suitable for network traffic"""
        input_layer = Input(shape=(input_dim,))
        
        # Encoder
        encoded = Dense(32, activation='relu')(input_layer)
        encoded = Dropout(0.2)(encoded)
        encoded = Dense(16, activation='relu')(encoded)
        encoded = Dropout(0.2)(encoded)
        encoded = Dense(8, activation='relu')(encoded)
        
        # Decoder
        decoded = Dense(16, activation='relu')(encoded)
        decoded = Dropout(0.2)(decoded)
        decoded = Dense(32, activation='relu')(decoded)
        decoded = Dropout(0.2)(decoded)
        decoded = Dense(input_dim, activation='sigmoid')(decoded)
        
        autoencoder = Model(input_layer, decoded)
        autoencoder.compile(optimizer='adam', loss='mse')
        return autoencoder
    
    def fit_models(self, X_train):
        """Fit all models with optimized parameters"""
        # Train Autoencoder
        self.autoencoder = self.build_autoencoder(X_train.shape[1])
        history = self.autoencoder.fit(
            X_train, X_train,
            epochs=100,
            batch_size=32,
            validation_split=0.2,
            callbacks=[
                tf.keras.callbacks.EarlyStopping(
                    patience=5,
                    restore_best_weights=True
                )
            ],
            verbose=1
        )
        
        # Train One-Class SVM
        self.ocsvm = OneClassSVM(
            kernel='rbf',
            nu=0.1,
            gamma='scale'
        ).fit(X_train)
        
        # Train Isolation Forest
        self.iforest = IsolationForest(
            n_estimators=200,
            contamination=0.1,
            random_state=42,
            max_samples='auto'
        ).fit(X_train)
        
        return history
    
    def predict_all(self, X):
        """Get predictions from all models"""
        # Autoencoder predictions
        reconstructed = self.autoencoder.predict(X)
        mse = np.mean(np.power(X - reconstructed, 2), axis=1)
        self.threshold = np.percentile(mse, 95)
        ae_pred = (mse > self.threshold).astype(int)
        
        # One-Class SVM predictions
        ocsvm_pred = (self.ocsvm.predict(X) == -1).astype(int)
        
        # Isolation Forest predictions
        iforest_pred = (self.iforest.predict(X) == -1).astype(int)
        
        self.predictions = {
            'Autoencoder': ae_pred,
            'One-Class SVM': ocsvm_pred,
            'Isolation Forest': iforest_pred
        }
        
        return self.predictions
    
    def evaluate_models(self, X_test, y_test):
        """Evaluate all models and store metrics"""
        predictions = self.predict_all(X_test)
        self.y_test = y_test
        
        for model_name, y_pred in predictions.items():
            self.metrics[model_name] = {
                'precision': precision_score(y_test, y_pred),
                'recall': recall_score(y_test, y_pred),
                'f1': f1_score(y_test, y_pred),
                'confusion_matrix': confusion_matrix(y_test, y_pred)
            }
    
    def plot_results(self):
        """Generate comprehensive visualization of results"""
        # 1. Model Performance Comparison
        self._plot_model_comparison()
        
        # 2. ROC Curves
        self._plot_roc_curves()
        
        # 3. Confusion Matrices
        self._plot_confusion_matrices()
        
        # 4. Feature Importance
        self._plot_feature_importance()
    
    def _plot_model_comparison(self):
        """Plot model performance metrics comparison"""
        metrics_df = pd.DataFrame({
            model: {
                'Precision': metrics['precision'],
                'Recall': metrics['recall'],
                'F1-Score': metrics['f1']
            }
            for model, metrics in self.metrics.items()
        }).T
        
        plt.figure(figsize=(12, 6))
        metrics_df.plot(kind='bar', ylim=(0, 1))
        plt.title('Model Performance Comparison', fontsize=14)
        plt.xlabel('Models', fontsize=12)
        plt.ylabel('Score', fontsize=12)
        plt.legend(bbox_to_anchor=(1.05, 1))
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
    
    def _plot_roc_curves(self):
        """Plot ROC curves for all models"""
        plt.figure(figsize=(10, 6))
        
        for model_name, predictions in self.predictions.items():
            fpr, tpr, _ = roc_curve(self.y_test, predictions)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
        
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves Comparison')
        plt.legend(loc="lower right")
        plt.grid(True, alpha=0.3)
        plt.show()
    
    def _plot_confusion_matrices(self):
        """Plot confusion matrices for all models"""
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        
        for i, (model_name, metrics) in enumerate(self.metrics.items()):
            sns.heatmap(
                metrics['confusion_matrix'],
                annot=True,
                fmt='d',
                cmap='Blues',
                ax=axes[i]
            )
            axes[i].set_title(f'{model_name}\nConfusion Matrix')
            axes[i].set_xlabel('Predicted')
            axes[i].set_ylabel('Actual')
        
        plt.tight_layout()
        plt.show()
    
    def _plot_feature_importance(self):
        """Plot feature importance based on model predictions"""
        # Calculate feature importance for each model
        feature_importance = np.abs(self.autoencoder.get_weights()[0])
        
        plt.figure(figsize=(12, 6))
        plt.bar(range(len(feature_importance)), np.mean(feature_importance, axis=1))
        plt.title('Feature Importance (based on Autoencoder weights)')
        plt.xlabel('Feature Index')
        plt.ylabel('Importance Score')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

def main():
    # Load your data
    df = df1_analyze
    
    # Initialize processors
    processor = NetworkFeatureProcessor()
    
    # Process features
    X, y = processor.process_features(df)
    X_processed = processor.scale_features(X)
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_processed, y,
        test_size=0.2,
        random_state=42,
        stratify=y
    )
    
    # Initialize and train detector
    detector = DDoSDetector()
    detector.fit_models(X_train)
    
    # Evaluate models
    detector.evaluate_models(X_test, y_test)
    
    # Plot results
    detector.plot_results()
    
    # Save models
    joblib.dump(detector, 'ddos_detector.joblib')
    joblib.dump(processor, 'feature_processor.joblib')

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import warnings
np.warnings = warnings
from sklearn.preprocessing import StandardScaler, PowerTransformer, LabelEncoder
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
import joblib
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

tf.config.threading.set_intra_op_parallelism_threads(os.cpu_count())
tf.config.threading.set_inter_op_parallelism_threads(os.cpu_count())

# Network Feature Processor (Unchanged)
class NetworkFeatureProcessor:
    """Class for processing network traffic features"""
    
    def __init__(self):
        self.label_encoder = LabelEncoder()
        self.scaler = StandardScaler()
        self.power_transformer = PowerTransformer()
    
    def process_features(self, df):
        """Process and transform network traffic features"""
        df = df.copy()
        
        # Encode Protocol
        df['Protocol'] = self.label_encoder.fit_transform(df['Protocol'])
        
        # Feature Grouping
        port_features = ['Source Port', 'Destination Port']
        packet_length_features = ['Fwd Packet Length Max', 'Fwd Packet Length Min', 
                                'Bwd Packet Length Min', 'Min Packet Length', 'Max Packet Length']
        timing_features = ['Bwd IAT Total', 'Bwd IAT Mean', 'Fwd Packets/s']
        flag_features = ['Fwd PSH Flags', 'SYN Flag Count', 'PSH Flag Count', 
                        'ACK Flag Count', 'URG Flag Count']
        other_features = ['Down/Up Ratio', 'Init_Win_bytes_backward', 
                         'min_seg_size_forward', 'Protocol']
        
        # Calculate additional features
        df['Length_Range'] = df['Max Packet Length'] - df['Min Packet Length']
        df['Flag_Sum'] = df[flag_features].sum(axis=1)
        
        # Combine all features
        feature_columns = (port_features + packet_length_features + timing_features + 
                         flag_features + other_features + 
                         [ 'Length_Range', 'Flag_Sum'])
        
        X = df[feature_columns].copy()
        y = (df['Label'] == 'DDoS').astype(int)  # Assuming 'DDoS' is the attack label
        
        return X, y
    
    def scale_features(self, X):
        """Scale and transform features"""
        # Apply log transformation to skewed features
        skewed_features = ['Bwd IAT Total', 'Bwd IAT Mean', 'Fwd Packets/s']
        for feature in skewed_features:
            if feature in X.columns:
                X[feature] = np.log1p(X[feature])
        
        # Standard scaling
        X_scaled = self.scaler.fit_transform(X)
        
        # Power transformation
        X_transformed = self.power_transformer.fit_transform(X_scaled)
        
        return X_transformed

# DDoS Detector with Hyperparameter Tuning and Autoencoder Improvements
class DDoSDetector:
    def __init__(self):
        self.autoencoder = None
        self.ocsvm = None
        self.iforest = None
        self.threshold = None
        self.metrics = {}
        self.predictions = {}
        
    def build_autoencoder(self, input_dim, encoding_dim=16, dropout_rate=0.2):
        """Build autoencoder with tunable architecture"""
        input_layer = Input(shape=(input_dim,))
        
        # Encoder
        encoded = Dense(encoding_dim, activation='relu')(input_layer)
        encoded = Dropout(dropout_rate)(encoded)
        encoded = Dense(encoding_dim // 2, activation='relu')(encoded)
        encoded = Dropout(dropout_rate)(encoded)
        
        # Decoder
        decoded = Dense(encoding_dim // 2, activation='relu')(encoded)
        decoded = Dropout(dropout_rate)(decoded)
        decoded = Dense(encoding_dim, activation='relu')(decoded)
        decoded = Dropout(dropout_rate)(decoded)
        decoded = Dense(input_dim, activation='sigmoid')(decoded)
        
        autoencoder = Model(input_layer, decoded)
        autoencoder.compile(optimizer='adam', loss='mse')
        return autoencoder
    
    def fit_autoencoder(self, X_train, epochs=100, batch_size=32, patience=5):
        """Fit Autoencoder with EarlyStopping"""
        self.autoencoder = self.build_autoencoder(X_train.shape[1])
        history = self.autoencoder.fit(
            X_train, X_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_split=0.2,
            callbacks=[tf.keras.callbacks.EarlyStopping(patience=patience, restore_best_weights=True)],
            verbose=1
        )
        return history
    
    def fit_models(self, X_train):
        """Fit all models with optimized parameters"""
        # Fit Autoencoder with Tuning
        self.fit_autoencoder(X_train)

        # Fit One-Class SVM with hyperparameter tuning
        self.ocsvm = self.optimize_ocsvm(X_train)
        
        # Fit Isolation Forest with hyperparameter tuning
        self.iforest = self.optimize_iforest(X_train)
        
    def optimize_ocsvm(self, X_train):
        """Optimize One-Class SVM hyperparameters using GridSearchCV"""
        param_grid = {
            'nu': [0.05, 0.1, 0.2],
            'gamma': ['scale', 'auto', 0.1],
            'kernel': ['rbf', 'poly']
        }
        ocsvm = OneClassSVM()
        grid_search = GridSearchCV(ocsvm, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
        grid_search.fit(X_train)
        return grid_search.best_estimator_

    def optimize_iforest(self, X_train):
        """Optimize Isolation Forest hyperparameters using GridSearchCV"""
        param_grid = {
            'n_estimators': [100, 200, 300],
            'contamination': [0.01, 0.05, 0.1],
            'max_samples': ['auto', 0.8]
        }
        iforest = IsolationForest()
        grid_search = GridSearchCV(iforest, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
        grid_search.fit(X_train)
        return grid_search.best_estimator_

    def predict_all(self, X):
        """Get predictions from all models"""
        # Autoencoder predictions
        reconstructed = self.autoencoder.predict(X)
        mse = np.mean(np.power(X - reconstructed, 2), axis=1)
        self.threshold = np.percentile(mse, 95)
        ae_pred = (mse > self.threshold).astype(int)
        
        # One-Class SVM predictions
        ocsvm_pred = (self.ocsvm.predict(X) == -1).astype(int)
        
        # Isolation Forest predictions
        iforest_pred = (self.iforest.predict(X) == -1).astype(int)
        
        self.predictions = {
            'Autoencoder': ae_pred,
            'One-Class SVM': ocsvm_pred,
            'Isolation Forest': iforest_pred
        }
        
        return self.predictions
    
    def evaluate_models(self, X_test, y_test):
        """Evaluate all models and store metrics"""
        predictions = self.predict_all(X_test)
        self.y_test = y_test
        
        for model_name, y_pred in predictions.items():
            self.metrics[model_name] = {
                'precision': precision_score(y_test, y_pred),
                'recall': recall_score(y_test, y_pred),
                'f1': f1_score(y_test, y_pred),
                'confusion_matrix': confusion_matrix(y_test, y_pred)
            }

    def plot_results(self):
        """Generate comprehensive visualization of results"""
        # 1. Model Performance Comparison
        self._plot_model_comparison()
        
        # 2. ROC Curves
        self._plot_roc_curves()
        
        # 3. Confusion Matrices
        self._plot_confusion_matrices()
    
    def _plot_model_comparison(self):
        """Plot model performance metrics comparison"""
        metrics_df = pd.DataFrame({
            model: {
                'Precision': metrics['precision'],
                'Recall': metrics['recall'],
                'F1-Score': metrics['f1']
            }
            for model, metrics in self.metrics.items()
        }).T
        
        plt.figure(figsize=(12, 6))
        metrics_df.plot(kind='bar', ylim=(0, 1))
        plt.title('Model Performance Comparison', fontsize=14)
        plt.xlabel('Models', fontsize=12)
        plt.ylabel('Score', fontsize=12)
        plt.legend(bbox_to_anchor=(1.05, 1))
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    def _plot_roc_curves(self):
        """Plot ROC curves for all models"""
        plt.figure(figsize=(10, 6))
        
        for model_name, predictions in self.predictions.items():
            fpr, tpr, _ = roc_curve(self.y_test, predictions)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
        
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves Comparison')
        plt.legend(loc="lower right")
        plt.grid(True, alpha=0.3)
        plt.show()

def main():
    # Load your data
    df = df1_analyze
    
    # Initialize processors
    processor = NetworkFeatureProcessor()
    
    # Process features
    X, y = processor.process_features(df)
    X_processed = processor.scale_features(X)
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_processed, y,
        test_size=0.2,
        random_state=42,
        stratify=y
    )
    
    # Initialize and train detector with hyperparameter tuning
    detector = DDoSDetector()
    detector.fit_models(X_train)
    
    # Evaluate models
    detector.evaluate_models(X_test, y_test)
    
    # Plot results
    detector.plot_results()
    
    # Save models
    joblib.dump(detector, 'ddos_detector.joblib')
    joblib.dump(processor, 'feature_processor.joblib')

if __name__ == "__main__":
    main()
```
